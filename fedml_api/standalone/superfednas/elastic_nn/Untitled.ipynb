```python
from darts.model import NetworkCIFAR
from darts import genotypes
```

```python
params = dict()
params["C"] = 24
params["num_classes"] = 10
params["layers"] = 4
params["auxiliary"] = False
params["genotype"] = genotypes.FedNAS_V1
model = NetworkCIFAR(**params)
```

```python
from torch import nn
import torch
```

```python
for m in model.modules():
    if isinstance(m, nn.BatchNorm2d) or isinstance(m, nn.BatchNorm1d):
        m.eval()
        m.weight.requires_grad = False
        m.bias.requires_grad = False
        m.running_mean.requires_grad = False
        m.running_var.requires_grad = False
        with torch.no_grad():
            m.weight.fill_(1)
            m.bias.fill_(0)
            m.running_mean.fill_(0)
            m.running_var.fill_(1)
        print(m.running_var)
```
